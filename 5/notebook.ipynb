```python
input = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""
```

```python
with open("input.txt", "r") as fp:
    input = fp.read().strip()
```

```python
data = input.split("\n")
i = data.index("")
ranges, ids = data[:i], data[i + 1 :]
ranges = [[int(n) for n in r.split("-")] for r in ranges]
ranges = sorted(ranges, key=lambda r: r[0])


def merge(ranges):
    new_ranges = [ranges[0]]

    for low, high in ranges:
        prev_low, prev_high = new_ranges[-1]
        if prev_high > high:
            continue

        if low - 1 <= prev_high:
            new_ranges[-1] = prev_low, high
            continue

        new_ranges.append((low, high))

    return new_ranges


ranges = merge(ranges)
```

```python
def is_fresh(id, min, max):
    return int(min) <= int(id) <= int(max)

sum([sum([is_fresh(id, *r) for r in ranges]) for id in ids])
```
